## <b>Introduction</b>: 

Toronto has the second-highest percentage of constant foreign-born population among world cities after Miami, Florida. Its diversity can be proven in the ethnic neighbourhoods, like Chinatown, Corso Italia, Greektown and etc. Chinese is reported to be one of the most common ethnic origins who live in Toronto, and thus that would be a huge potential in launching a Chinese restaurant.

This project is aimed to help those entrepreneurs to identifying which neighbourhood in Toronto is the best neighbourhood to have a Chinese restaurant.  


## Target Audience: 
The target audience of this report would be focused on the category whose are interested in having a Chinese restaurant business or to grow their business. This report will provide them some useful information that can be applied as a reference or a suggestion.
